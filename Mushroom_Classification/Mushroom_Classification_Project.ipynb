```python
#stalk-shape(菌柄)：enlarging：菌柄自基部向上逐渐变宽，tapering：菌柄自基部向上逐渐变细

#stalk-root:b (bulbous)：菌柄根部呈膨大状,c (club)：菌柄根部呈棍棒状,e (equal)：菌柄上下一致,r (rooted)：菌柄有一个明显的根状结构,? (missing)

#stalk-surface-above(below)-ring菌柄环上方表面的特征:f (fibrous)：纤维状的表面，y (scaly)：鳞片状的表面，k (silky)：丝质的表面，s (smooth)：光滑的表面

#stalk-color-above(below)-ring:n (brown)：棕色,b (buff)：浅黄色（浅褐色）,c：肉桂色, g：灰色, o：橙色, p：粉红色, e：红色,w (white)：白色,y (yellow)：黄色

#veil-type菌幕类型：partial：部分菌幕

#veil-color:n (brown)：棕色,o (orange)：橙色,w (white)：白色,y (yellow)：黄色

#ring-number菌环数量：n (none)：没有菌环,o (one)：有一个菌环。,t (two)：有两个菌环。(需使用标签编码)

#ring-type：e (evanescent)：易消失或不明显。f (flaring)：向外扩展的，边缘向外翻开。l (large)：菌环较大。n (none)：没有.p (pendant)：呈下垂状，像垂下来的环

#spore-print-color蘑菇孢子印颜色：k (black),n (brown),b (buff),h (chocolate),r (green),o (orange),u (purple),w (white),y (yellow)

#population:a (abundant)：丰富的，蘑菇大量生长。c (clustered)：成簇生长，蘑菇紧密聚集在一起。n (numerous)：数量众多，生长得比较多，但不像成簇那样紧密。
#s (scattered)：分散生长，蘑菇不密集地分布。v (several)：数个，蘑菇少量生长。y (solitary)：独生，通常单个生长

#habitat:g (grasses)：草地，蘑菇通常生长在草丛中。l (leaves)：林地，蘑菇生长在树叶覆盖的地方。m (meadows)：草甸，蘑菇生长在开阔的草地或草原。
#p (paths)：小路旁，蘑菇沿小路或人类活动区边缘生长。u (urban)：城市区域，蘑菇生长在人类居住地、城市公园或建筑附近。
#w (waste)：荒地，蘑菇生长在荒废或垃圾堆积的地方。d (woods)：森林，蘑菇生长在树木茂密的森林中。
```

### 特征工程-聚类

```python
# 提取低相关性特征名
low_corr_columns = low_corr_features.index.tolist()

# 创建分组字典
feature_groups = {
    'cap': [],
    'stalk': [],
    'gill':[],
    'spore':[],
    'other': []
}

# 遍历低相关性特征，根据特征名的前缀进行分组
for feature in low_corr_columns:
    if feature.startswith('cap'):
        feature_groups['cap'].append(feature)
    elif feature.startswith('stalk'):
        feature_groups['stalk'].append(feature)
    elif feature.startswith('gill'):
        feature_groups['gill'].append(feature) 
    elif feature.startswith('spore'):
        feature_groups['spore'].append(feature)  
    else:
        feature_groups['other'].append(feature)

# 输出分组后的特征
print("Cap features:", feature_groups['cap'])
print("Stalk features:", feature_groups['stalk'])
print("Gill features:", feature_groups['gill'])
print("Spore features:", feature_groups['spore'])
print("Other features:", feature_groups['other'])
```

```python
print(feature_groups.items())
```

```python
def perform_kmeans_clustering(mc, features, n_clusters=3):
    # 提取特征数据
    subset = mc[features]
    
    # 进行 KMeans 聚类
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(subset)

    # 将聚类结果添加到数据集中
    mc[f'{features[0]}_cluster'] = kmeans.labels_  # 使用第一个特征名作为后缀

    # 查看聚类结果
    print(mc[[f'{features[0]}_cluster'] + features].head())
    
    return mc

# 调用方法只对 'cap' 组聚类
mc = perform_kmeans_clustering(mc, feature_groups['cap'], n_clusters=3)
```

```python
def visualize_kmeans(mc, features, cluster_column):
    # 提取特征数据
    subset = mc[features]

    # 使用 PCA 降维到 2D
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(subset)
    
    # 将 PCA 结果添加到数据集中
    mc['pca_one'] = pca_result[:, 0]
    mc['pca_two'] = pca_result[:, 1]

    # 可视化聚类结果
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        x='pca_one', 
        y='pca_two', 
        hue=cluster_column, 
        data=mc, 
        palette='viridis',
        legend='full',
        alpha=0.7
    )
    
    plt.title('KMeans Clustering Visualization')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend(title=cluster_column)
    plt.show()
```
